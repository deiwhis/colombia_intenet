```python
# Eliminar espacios extra en los nombres de columnas
accesos_fid.columns = accesos_fid.columns.str.strip()
```

```python
# Eliminar puntos y convertir a int64
accesos_fid["No. ACCESOS FIJOS A INTERNET"] = accesos_fid["No. ACCESOS FIJOS A INTERNET"].str.replace('.', '', regex=False).astype(int)
accesos_fid["POBLACIÓN DANE"] = accesos_fid["POBLACIÓN DANE"].str.replace('.', '', regex=False).astype(int)
```

```python
column_rename_map = {"No. ACCESOS FIJOS A INTERNET": "numero_accesos_fijos", "POBLACIÓN DANE":"poblacion_dane", "DEPARTAMENTO": "departamento", "AÑO": "año", "TRIMESTRE": "trimestre", "CÓDIGO DANE": "codigo_dane", "PENETRACIÓN": "penetracion"}
```

```python
accesos_fid_renombrado = accesos_fid.rename(columns=column_rename_map)
```

```python
accesos_fid_renombrado.head()
```

```python
# Filtrar datos del trimestre 3 de los años 2023 y 2024
accesos_filtrados = accesos_fid[(accesos_fid["trimestre"] == 3) & (accesos_fid["año"].isin([2023, 2024]))]

# Agrupar por año y sumar accesos
accesos_por_año = accesos_filtrados.groupby("año")["numero_accesos_fijos"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ax = plt.bar(accesos_por_año["año"].astype(str), accesos_por_año["numero_accesos_fijos"],
             color=["skyblue", "steelblue"], edgecolor="black")

# Anotar valores exactos
for bar in ax:
    height = bar.get_height()
    plt.annotate(f'{height:,}',  # formato con comas
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5), textcoords="offset points",
                 ha='center', va='bottom', fontsize=12)

plt.title("Gráfico 1: Total de accesos fijos a Internet en Colombia\nTercer trimestre 2023 vs. 2024", fontsize=14)
plt.xlabel("Año")
plt.ylabel("Número de Accesos Fijos")
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Agrupar por año y trimestre y sumar accesos
serie_trimestral = accesos_fid.groupby(["año", "trimestre"])["numero_accesos_fijos"].sum().reset_index()

# Crear una columna con etiquetas tipo "I trim. 2022", "III trim. 2024", etc.
mapa_trimestres = {1: "I trim.", 2: "II trim.", 3: "III trim.", 4: "IV trim."}
serie_trimestral["etiqueta"] = serie_trimestral.apply(
    lambda row: f'{mapa_trimestres[row["trimestre"]]} {row["año"]}', axis=1)

# Ordenar cronológicamente
serie_trimestral = serie_trimestral.sort_values(by=["año", "trimestre"])
```

```python
# Crear una nueva columna con valores en millones (float)
serie_trimestral["millones"] = serie_trimestral["numero_accesos_fijos"] / 1_000_000
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(serie_trimestral["etiqueta"], serie_trimestral["millones"], color="#F4A300", marker='o', linewidth=2)

# Anotar valores sobre cada punto
for x, y in zip(serie_trimestral["etiqueta"], serie_trimestral["millones"]):
    plt.annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0, 6), ha='center')

plt.title("Gráfico 1. Accesos fijos a Internet (millones)", fontsize=16, color="#002f6c", weight='bold')
plt.ylabel("Accesos (millones)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Agrupar por departamento, año y trimestre
serie_dep = accesos_fid.groupby(["departamento", "año", "trimestre"])["numero_accesos_fijos"].sum().reset_index()

# Crear etiquetas como "III trim. 2024"
mapa_trimestres = {1: "I trim.", 2: "II trim.", 3: "III trim.", 4: "IV trim."}
serie_dep["etiqueta"] = serie_dep.apply(
    lambda row: f'{mapa_trimestres[row["trimestre"]]} {row["año"]}', axis=1)

# Ordenar por año y trimestre
serie_dep = serie_dep.sort_values(by=["departamento", "año", "trimestre"])
```

```python
# Seleccionar los 5 con más accesos en total (opcional)
top_departamentos = serie_dep.groupby("departamento")["numero_accesos_fijos"].sum().nlargest(5).index.tolist()

# Filtrar solo esos departamentos
serie_top = serie_dep[serie_dep["departamento"].isin(top_departamentos)].copy()

# Convertir a millones
serie_top["millones"] = serie_top["numero_accesos_fijos"] / 1_000_000
```

```python
plt.figure(figsize=(16, 8))

for dep in top_departamentos:
    datos_dep = serie_top[serie_top["departamento"] == dep]
    plt.plot(datos_dep["etiqueta"], datos_dep["millones"], marker='o', label=dep)

plt.title("Accesos fijos a Internet por Departamento (millones)", fontsize=15, color="#002f6c", weight='bold')
plt.xlabel("Trimestre")
plt.ylabel("Accesos (millones)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Departamento")
plt.tight_layout()
plt.show()
```

```python
# Filtrar trimestre 3 de 2024
accesos_t3_2024 = accesos_fid[(accesos_fid["año"] == 2024) & (accesos_fid["trimestre"] == 3)]

# Agrupar por departamento
resumen_departamentos = accesos_t3_2024.groupby("departamento")["numero_accesos_fijos"].sum().reset_index()

# Ordenar por mayor cantidad
resumen_departamentos = resumen_departamentos.sort_values(by="numero_accesos_fijos", ascending=False)
```

```python
import matplotlib.pyplot as plt

# Paso 1: Filtrar el III trimestre de 2024
accesos_t3_2024 = accesos_fid[(accesos_fid["año"] == 2024) & (accesos_fid["trimestre"] == 3)]

# Paso 2: Agrupar por departamento y ordenar de mayor a menor
resumen_departamentos = accesos_t3_2024.groupby("departamento")["numero_accesos_fijos"].sum().reset_index()
resumen_departamentos = resumen_departamentos.sort_values(by="numero_accesos_fijos", ascending=False)

# Paso 3: Crear gráfico de barras verticales
plt.figure(figsize=(16, 8))  # Ancho suficiente para mostrar etiquetas
bars = plt.bar(resumen_departamentos["departamento"], resumen_departamentos["numero_accesos_fijos"],
               color="cornflowerblue", edgecolor="black")

# Anotar valores encima de las barras
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:,}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5), textcoords="offset points",
                 ha='center', va='bottom', fontsize=8)

plt.title("Accesos fijos a Internet por Departamento\nIII trim. 2024", fontsize=16, color="#002f6c", weight='bold')
plt.xlabel("Departamento")
plt.ylabel("Número de Accesos Fijos")
plt.xticks(rotation=90, ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```
