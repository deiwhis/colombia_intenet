```python
# Eliminar espacios extra en los nombres de columnas
accesos_fid.columns = accesos_fid.columns.str.strip()
```

```python
# Eliminar puntos y convertir a int64
accesos_fid["No. ACCESOS FIJOS A INTERNET"] = accesos_fid["No. ACCESOS FIJOS A INTERNET"].str.replace('.', '', regex=False).astype(int)
accesos_fid["POBLACIÓN DANE"] = accesos_fid["POBLACIÓN DANE"].str.replace('.', '', regex=False).astype(int)
```

```python
column_rename_map = {"No. ACCESOS FIJOS A INTERNET": "numero_accesos_fijos", "POBLACIÓN DANE":"poblacion_dane", "DEPARTAMENTO": "departamento", "AÑO": "año", "TRIMESTRE": "trimestre", "CÓDIGO DANE": "codigo_dane", "PENETRACIÓN": "penetracion"}
```

```python
accesos_fid_renombrado = accesos_fid.rename(columns=column_rename_map)
```

```python
accesos_fid_renombrado.head()
```

```python
# Filtrar datos del trimestre 3 de los años 2023 y 2024
accesos_filtrados = accesos_fid[(accesos_fid["trimestre"] == 3) & (accesos_fid["año"].isin([2023, 2024]))]

# Agrupar por año y sumar accesos
accesos_por_año = accesos_filtrados.groupby("año")["numero_accesos_fijos"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ax = plt.bar(accesos_por_año["año"].astype(str), accesos_por_año["numero_accesos_fijos"],
             color=["skyblue", "steelblue"], edgecolor="black")

# Anotar valores exactos
for bar in ax:
    height = bar.get_height()
    plt.annotate(f'{height:,}',  # formato con comas
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5), textcoords="offset points",
                 ha='center', va='bottom', fontsize=12)

plt.title("Gráfico 1: Total de accesos fijos a Internet en Colombia\nTercer trimestre 2023 vs. 2024", fontsize=14)
plt.xlabel("Año")
plt.ylabel("Número de Accesos Fijos")
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Agrupar por año y trimestre y sumar accesos
serie_trimestral = accesos_fid.groupby(["año", "trimestre"])["numero_accesos_fijos"].sum().reset_index()

# Crear una columna con etiquetas tipo "I trim. 2022", "III trim. 2024", etc.
mapa_trimestres = {1: "I trim.", 2: "II trim.", 3: "III trim.", 4: "IV trim."}
serie_trimestral["etiqueta"] = serie_trimestral.apply(
    lambda row: f'{mapa_trimestres[row["trimestre"]]} {row["año"]}', axis=1)

# Ordenar cronológicamente
serie_trimestral = serie_trimestral.sort_values(by=["año", "trimestre"])
```

```python
# Crear una nueva columna con valores en millones (float)
serie_trimestral["millones"] = serie_trimestral["numero_accesos_fijos"] / 1_000_000
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(serie_trimestral["etiqueta"], serie_trimestral["millones"], color="#F4A300", marker='o', linewidth=2)

# Anotar valores sobre cada punto
for x, y in zip(serie_trimestral["etiqueta"], serie_trimestral["millones"]):
    plt.annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0, 6), ha='center')

plt.title("Gráfico 1. Accesos fijos a Internet (millones)", fontsize=16, color="#002f6c", weight='bold')
plt.ylabel("Accesos (millones)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
